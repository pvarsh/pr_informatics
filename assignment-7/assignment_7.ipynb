```python
import numpy as np
from matplotlib import pyplot as plt
import pylab
import pandas as pd
from datetime import datetime
import sys
import matplotlib.cm as cm

### settings
%matplotlib inline
```

###### Functions

```python
def top_k_types(df, k):
    grouped = df['Complaint Type'].groupby(df['Complaint Type'])
    complaints = grouped.count().sort(inplace = False, ascending = False)
    return complaints[0:k]
    
```

```python
### read data
### reading is slow due to parsing two columns as datetime
in_file = "311_Service_Requests_from_2010_to_Present.csv"
with open(in_file, 'r') as f:
    df = pd.read_csv(f, index_col = ['Unique Key'], parse_dates=['Created Date', 'Closed Date'])
```

```python
datemin = min(df['Created Date'])
datemax = max(df['Created Date'])
```

```python
datemin_str = datetime.strftime(datemin, "%b/%d/%Y")
datemax_str = datetime.strftime(datemax, "%b/%d/%Y")
```

```python
grouped = df['Agency'].groupby(df['Agency'])
pr1_1 = grouped.count()
```

```python
### Problem 1
pr1_agencies = ['NYPD','DOT','DOB','TLC','DPR']
```

```python
#colors = [0.5/float(len(pr1_agencies)) * k for k in range(len(pr1_agencies))]
#colors = np.random.randn(len(pr1_agencies))
fig, ax = plt.subplots()
xpos = range(len(pr1_agencies))
plt.bar(xpos, pr1_1[pr1_agencies], align = 'center', color = cm.Blues(0.4))
ax.set_xticks(xpos)
ax.set_xlabel("Agency")
yticks = [int(x) for x in ax.get_yticks()]
ax.set_yticklabels(yticks, fontsize = 8)
ax.set_ylabel("Volume")
ax.set_title("Complaint volumes by agency between %s - %s" %(datemin_str, datemax_str), fontsize = 9)
ax.set_xticklabels( pr1_agencies, fontsize = 9)
plt.show()
```

###### Problem 2

```python
k = 6
complaint_types = top_k_types(df, k)
print list(complaint_types.keys())
fig, ax = plt.subplots(figsize = (10, 5))
xpos = [1*x for x in range(len(complaint_types))]
plt.bar(xpos, complaint_types, align = 'center', width = .8, color = cm.Blues(0.4))
ax.set_title("Top %d complaint types %s-%s" %(k, datemin_str, datemax_str))
ax.set_xticks(xpos)
lbls = list(complaint_types.keys())
ax.set_xticklabels(complaint_types.keys(), fontsize = 7)
ax.set_xlabel("Complaint Type")
ax.set_ylabel("Volume")
plt.show()
```
