```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import rc

%matplotlib inline
```

```python
datafile = "microprocessors.dat"
with open(datafile, 'r') as f:
    procs = pd.read_csv(datafile)
```

```python
procs['log_trans'] = np.log10(procs['Transistors'])
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
ax.plot(procs['Year of Introduction'], procs.log_trans, 'or')
for ind in procs.index:
    xy = (procs['Year of Introduction'][ind] +.3, procs.log_trans[ind] - .3)
    ax.annotate(procs.Processor[ind], xy = xy, fontsize = 10)
### axis labels
ax.set_xlabel('Year')
ax.set_ylabel('Number of transistors')
### ticks
yticks = [int(tick) for tick in ax.get_yticks()]
yticks = [r'$10^{%d}$' %tick for tick in yticks]
ax.set_yticklabels(yticks, fontsize = 12)
plt.show()
```
